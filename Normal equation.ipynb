```python
X = data.iloc[:,0:-1]#切片取第一列到倒数第二列构建数据集X
X.head()
```

```python
y=data.iloc[:,-1]#切片取倒数第二列到倒数第一列构建标签y
y.head()
```

```python
X=X.values#将DataFrame格式转换为数组/矩阵
X.shape#检查维度
```

```python
y=y.values#将DataFrame格式转换为数组/矩阵
y.shape
```

```python
y=y.reshape(97,1)#将y转化为二维数组，便于后续运算
y.shape
```

```python
def normalEquation(X,y):#构建正规方程
    theta=np.linalg.inv(X.T @ X) @ X.T @ y#inalg.inv矩阵求逆
    return theta
```

```python
theta = normalEquation(X,y)
print(theta)
```
